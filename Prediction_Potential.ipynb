### Scikit-Learn

```python
#librer para analisis 
#sklearn extrae data de entrenemiento y de testeo
#libera train_test_split ayuda a separar los datos

from sklearn.model_selection import train_test_split #ayuda a separar los datos
```

```python
# dividir para poder llenar las 4 variables
#test_size indica la proporcin de separacin para los datos de testeo
#x_train , x_test, y_train, y_test = variables devueltas por la funcion
x_train , x_test, y_train, y_test = train_test_split(df['Age'], df['Potential'], test_size = 0.2)
print(x_train , x_test, y_train, y_test)


```

```python
#la x entrenamiento la divide en 14565 filas --> jugadores para entrenar
#la x de testeo la divide en 3642 filas --> jugadores para testear
print(x_train.shape, x_test.shape)
print(type(x_train), type(x_test))
```

```python
# seleccionar un modelo de regrecin lineal
from sklearn.linear_model import LinearRegression
```

```python
#la clase lineal regresion ejecuto
#almaceno en una variable
regression = LinearRegression()
```
